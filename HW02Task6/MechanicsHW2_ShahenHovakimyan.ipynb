```python
import matplotlib.pyplot as plt

def simulate_motion(m1, m2, m3, mu1, mu2, mu3, L, dt):
  g = 9.81
  t = 0
  x1, x2, y3 = 0, L, 0
  v1, v2, v3 = 0, 0, 0
  positions = [[], [], []]
  times = []

  while y3 >= 0:
    a1 = (m3*g - mu3*m2*g) / (m1 + m2 + m3)
    a2 = -a1
    a3 = g - a1

    v1 += a1 * dt
    v2 += a2 * dt
    v3 += a3 * dt

    x1 += v1 * dt
    x2 += v2 * dt
    y3 += v3 * dt

    if x2 <= 0 or y3 <= 0:
      break

    t += dt
    positions[0].append(x1)
    positions[1].append(x2)
    positions[2].append(y3)
    times.append(t)

  return positions, times

def visualize_motion(positions, times):
  x1, x2, y3 = positions
  plt.plot(times, x1, label="M1")
  plt.plot(times, x2, label="M2")
  plt.plot(times, y3, label="M3")
  plt.xlabel("Time (s)")
  plt.ylabel("Position (m)")
  plt.legend()
  plt.grid(True)
  plt.show()

m1 = 10  # kg
m2 = 5  # kg
m3 = 8  # kg
mu1 = 0.2
mu2 = 0
mu3 = 0.1
L = 1.5  # m
dt = 0.01  # s

positions, times = simulate_motion(m1, m2, m3, mu1, mu2, mu3, L, dt)

max_distance = max(positions[0])

print("Maximum distance traveled by M1:", max_distance, "m")

visualize_motion(positions, times)
```
